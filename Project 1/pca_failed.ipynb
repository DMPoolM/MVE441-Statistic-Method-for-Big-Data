```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
x_1_dataset = np.zeros((50,2))
x_2_dataset= x_1_dataset.copy()
x_1 = np.random.normal(5,10,50)
x_2 = np.random.normal(5, 10, 50)

y_1 = x_1 *2 + 30 + np.random.random((50)) *10
y_2 = x_2 *2 +  np.random.random((50)) * 20

x_1_dataset[:, 0], x_1_dataset[:, 1] = x_1, y_1
class_1 = np.zeros((50,1))

x_2_dataset[:, 0], x_2_dataset[:, 1] = x_2, y_2
class_2 = np.ones((50,1))

X_dataset = np.append(x_1_dataset, x_2_dataset, axis=0)
y_dataset = np.append(class_1, class_2, axis=0)

dataset = np.append(X_dataset, y_dataset, axis=1)
plt.scatter(x_1, y_1, label="class1")
plt.scatter(x_2, y_2, label='class2')
plt.legend()
plt.show()

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_dataset, y_dataset, test_size=0.3)

reg_model = LogisticRegression().fit(X_train, y_train)
reg_model.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1, whiten=True, random_state=0).fit(X_train)
X_train_pca0 = pca.transform(X_train)
X_test_pca0 = pca.transform(X_test)


plt.scatter(X_train_pca0, [0] * X_train_pca0.shape[0], c=y_train)
plt.legend()

reg_model = LogisticRegression().fit(X_train_pca0, y_train)
reg_model.score(X_test_pca0, y_test)
```
